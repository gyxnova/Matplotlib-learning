## stack plot

```python
x=[1,2,3,4,5]
area=[1,2,3,4,5]
area1=[2,3,4,3,6]
area2=[3,4,2,6,7]
plt.stackplot(x,area,area1,area2,labels=["python","c","c++"],baseline="zero")
plt.title("Stack plot")
plt.xlabel("x")
plt.ylabel("y")
plt.savefig("stack plot",facecolor="m")
plt.legend()
plt.show()
```

## Step plot

```python
x= [1,2,3,4,5,6]
y=[4,5,7,6,8,5]
plt.step(x,y,where="post",label="step plot",marker = "o",ms = 10,mfc="g")
plt.title("Step plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
